# Present Continuous
---

| Affirmative                        | Negative                                     | Questions         |
| --                                 | --                                           | --                |
| I am waiting.     I'm waiting.     | I am not waiting.     I'm not waiting.       | Am I waiting?     |
| You are waiting.  You're waiting.  | You are not waiting.  You aren't waiting.    | Are you waiting?  |
| He is waiting.    He's waiting.    | He is not waiting.    He isn't waiting.      | Is he waiting?    |
| We are waiting.   We're waiting.   | We are not waiting.   We aren't waiting.     | Are we waiting?   |
| They are waiting. They're waiting. | They are not waiting. They aren't waiting.   | Are they waiting? |
